**Question 3.** Specify the independent variable and blocking factor. What are some possible lurking variables?

The independent variable would be the categorical variable Teaching_Method with 3 levels. The blocking factor in this experiment would be the variable Prior_Knowledge. This variable has two levels High and Low distinguishing the intellectual abilities of certain undergraduate students. **Question 4.** Briefly explain why a randomized block design would be appropriate here. Similarly, explain why a completely randomized design would not be appropriate.

This lets us know that we are working with heterogeneous data in respect to student's prior knowledge. If we used a Completely Randomized design the outcome would be inaccurate due to the sheer fact that we have a present lurking variable Prior_Knowledge. We would need to control for this variable while analyzing the treatment variable Teaching_Method to ensure we have a reputable outcome.
